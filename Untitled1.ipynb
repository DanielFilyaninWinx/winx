```python
import numpy as np                   
import matplotlib.pyplot as plt      
import random                        

from keras.datasets import mnist     
from keras.models import Sequential  

from keras.layers.core import Dense, Dropout, Activation 
from keras.utils import np_utils                         
```

# ***Подготавливаем данные для обучения и проверки***

подгружаем набор данных при помощи встроенных методов keras

```python

(X_train, y_train), (X_test, y_test) = mnist.load_data()#х_train - данные для обучения y_train - правильные ответы

```

Просматриваем примеры изображений

```python
plt.rcParams['figure.figsize'] = (9,9)
for i in range(9):
    plt.subplot(3,3,i+1)
    n = random.randint(0, len(X_train))
    plt.imshow(X_train[n], cmap='gray', interpolation='none')
    plt.title("Class {}".format(y_train[n]))
    
plt.tight_layout()
```

***Преобразование размерности данных в наборе***

```python
X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)
```

***Нормализация данных***

```python
X_train = X_train.astype('float32') 
X_test = X_test.astype('float32')

X_train /= 255    #делим интенсивность каждого пикселя изображения на 255                   
X_test /= 255
```

# ***Работа с правильными ответами***

```python
print(y_train[n])
```

```python
#Преобразуем метки в формат one hot encoding

Y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(y_test, 10)
```

```python
#Правильный ответ в формате one hot encoding
print(y_train[n])
```

# ***Создаем нейронную сеть***

***Создаем последовательную модель***

```python
model = Sequential()
```

***Добавляем уровни/слои сети***

```python
model.add(Dense(800, input_dim=784, activation="relu"))
model.add(Dense(10, activation="softmax"))
```

***Компилируем сеть***

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

print(model.summary())
```

# ***Обучаем нейронную сеть***

```python
model.fit(X_train, Y_train, batch_size=200, epochs=20,  verbose=1)
```

# ***Сохраняем обученную нейронную сеть***

***Генерируем описание модели в формате json***

```python
model_json = model.to_json()
```

```python
print(model_json)
```

***Записываем архитектуру сети в файл***

```python
json_file = open("mnist_model_two.json", "w")
json_file.write(model_json)
json_file.close()
```

***Записываем данные о весах сети в файл***

```python
model.save_weights("mnist_model_two.h5")#HDF5 позволяет хранить бинарные данные кот-е имеют сложную структурную сеть
```

Проверка качества работы на наборе данных для тестирования

```python
scores = model.evaluate(X_test, Y_test, verbose=1)
```
